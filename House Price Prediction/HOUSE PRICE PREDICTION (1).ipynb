```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=5)
```

```python
print(X_train.shape)
print(y_train.shape)
```

<h2>LINEAR REGRESSION</h2>

```python
'''make_column_transformer:
This function is used to create a transformer that can apply different preprocessing steps to
different columns of a dataset. It's particularly useful when you need to preprocess numerical
and categorical data differently.

The OneHotEncoder is used to convert categorical data into a numerical format that machine
learning algorithms can work with. Specifically, it converts categorical values into a one-hot
encoded format, where each category is represented by a binary vector.

sparse=False specifies that the output should be a dense array rather than a sparse matrix.
In a dense array, all values are stored explicitly, which is easier to inspect and work with
for small datasets.

['beds']:
This is the list of column(s) that the OneHotEncoder should be applied to. In this case, it's a
single column named 'beds'.
The 'beds' column likely contains categorical data, such as the number of beds in a house
(e.g., "1 bed," "2 beds"), which will be one-hot encoded.

remainder='passthrough':
This parameter specifies what to do with the columns that are not explicitly mentioned in the
transformer (i.e., columns other than 'beds').'passthrough' means that all other columns
should be left unchanged and passed through without any transformation.
'''
```

```python
cl_trans = make_column_transformer((OneHotEncoder(handle_unknown="ignore"), ['beds']), remainder='passthrough')
```

```python
'''The standardization process involves subtracting the mean of each feature and dividing by
the standard deviation:
After applying StandardScaler, the feature will have a mean of 0 and a standard deviation of 1.
If your dataset contains features with different scales (e.g., one feature in the range of 0-1
and another in the range of 0-1000), standardizing them ensures that the model treats them
equally.'''
```

```python
# X being feature matrix
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

lr = LinearRegression()
lr.fit(X_scaled, y)
```

```python
'''The pipeline ensures that these steps are executed in sequence: first, the data 
is transformed using cl_trans, then it is scaled using scaler, and finally,
the transformed and scaled data is passed to the lr model for fitting.'''
```

```python
pipe=make_pipeline(cl_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

<h2>LASSO</h2>

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(cl_trans,scaler, lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

<h2>RIDGE</h2>

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(cl_trans,scaler, ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test,y_pred_ridge)
```

```python
print("No Regularization: ", r2_score(y_test,y_pred_lr))
print("Lasso: ", r2_score(y_test,y_pred_lasso))
print("Ridge: ", r2_score(y_test,y_pred_ridge))
```
